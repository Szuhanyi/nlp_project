```python
import pandas as pd
import os
import string
import nltk
from nltk.corpus import stopwords
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB


from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import confusion_matrix
import numpy

```

```python
all_word_lexicon = {}
# nltk.download('stopwords')
stop_words = stopwords.words('english')

df = pd.read_csv("./input/train.csv")
for index, row in df.iterrows():
  sentiment = row['sentiment']
  review = row['review'].translate(str.maketrans('', '', string.punctuation))
  review = review.lower().split(" ")
  review = [word for word  in review if not word in stop_words]
  for word in review:
    if word not in all_word_lexicon.keys():
      all_word_lexicon[word] = 0
    if sentiment:
      all_word_lexicon[word] += 1
    else:
      all_word_lexicon[word] -= 1

df = pd.read_csv("./input/test.csv")
for index, row in df.iterrows():
  sentiment = row['sentiment']
  review = row['review'].translate(str.maketrans('', '', string.punctuation))
  review = review.lower().split(" ")
  review = [word for word  in review if not word in stop_words]
  for word in review:
    if word not in all_word_lexicon.keys():
      all_word_lexicon[word] = 0
    if sentiment:
      all_word_lexicon[word] += 1
    else:
      all_word_lexicon[word] -= 1

all_word_lexicon
```

```python
all_word_lexicon_sorted = sorted(all_word_lexicon.items(), key=lambda item: item[1])
positive = all_word_lexicon_sorted[:1000]
negative = all_word_lexicon_sorted[-1000:]
lexicon = dict(positive+negative)
print(len(lexicon))
lexicon
```

```python


df_model = pd.read_csv('./input/IMDB Dataset.csv')


X = df_model['review']
y = df_model['sentiment']
del df_model

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

vec = CountVectorizer(stop_words='english', min_df = 20)
vec.fit(X)

X_train_transformed = vec.transform(X_train)
X_test_transformed = vec.transform(X_test)

print(X_train_transformed.shape)
print(X_test_transformed.shape)
```

```python
model = BernoulliNB()
model.fit(X_train_transformed.toarray(),y_train)
del X_train_transformed
```

```python
predictions = model.predict(X_test_transformed.toarray())
```

```python
conf_matr = confusion_matrix(y_test, predictions)
```

```python
(conf_matr[0,0]+conf_matr[1,1])/10000

```

```python
prediction_with_lexicon = []
index = 0
for review in X_test:
    review = review.split(" ")
    review = [word for word  in review if not word in stop_words]
    # print(review)
    score = 0
    absolute_score = 0
    for word in review:
        if word in lexicon.keys():
            score += lexicon[word]
            absolute_score += abs(lexicon[word])
    if score == 0:
        value = 0
    else:
        value = (score/absolute_score + 1)/2
    if (pred_prob[index,1]+value)/2 > 0.5:
        prediction_with_lexicon.append("positive")
    else:
        prediction_with_lexicon.append("negative")
    index += 1

# df_pred_lex = pd.Series(prediction_with_lexicon)
# df_pred_lex.hist(bins= numpy.linspace(-1, 1, num=100))
```
